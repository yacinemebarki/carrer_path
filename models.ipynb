```python
import re

def clean_string(s):
    if isinstance(s, str):
        s=re.sub(r'[^\w\s]', ' ', s)
        s=re.sub(r'\s+', ' ', s).strip()
        return s
    return s
 
```

```python
career
```

```python
career["Skill"]=career["Skill"].apply(lambda x:clean_string(x))
career
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
```

```python
x=career["Skill"]
y=career["Career"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
tr=TfidfVectorizer()
x_vec=tr.fit_transform(x_train)
x_vec1=tr.transform(x_test)
```

```python
vec=[]
i=0
score=[]
for test in x_vec1:
    value=cosine_similarity(test,x_vec).flatten()
    position=value.argmax()
    vec.append(y_train.iloc[i])
    i=i+1
    score.append(value[position])
result=pd.DataFrame(vec)
result    

```

```python
print(score)
```

```python
print(sum(score)/len(score))
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report


y_true=y_test
y_pred=vec  
acc=accuracy_score(y_true, y_pred)
print("Accuracy:", acc)
f1=f1_score(y_true, y_pred, average="macro")
print("F1-score:", f1)
print(classification_report(y_true, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe = Pipeline([
    ("tfidf", TfidfVectorizer(stop_words="english")),
    ("clf", LogisticRegression(max_iter=1000))
])
pipe.fit(x_train, y_train)
y_pred=pipe.predict(x_test)
print(classification_report(y_test, y_pred))
print(accuracy_score(y_pred,y_test))
print(f1_score(y_pred,y_test,average="macro"))
```
