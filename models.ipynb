```python
student=student.drop(["Department","Gender"],axis=1)
print(student[student["Job"]=="Yes"].count())
```

```python
student=student.drop("Job",axis=1)
```

```python
print(student[student["Extra"]=="Yes"].count())
student=student.drop("Income",axis=1)
```

```python
print(student["Hometown"].unique())
```

```python
student["Hometown"]=student["Hometown"].replace({
    "Village":0,
    "City":1
})
student["Extra"]=student["Extra"].replace({
    "Yes":1,
    "No":0
})
student
```

```python
student["Semester"]=student["Semester"].str[0].astype(int)
student
```

```python
print(student["Attendance"].unique())
```

```python
student["Attendance"].unique()
```

```python
student["Attendance"]=student["Attendance"].replace({
    "Below 40":"0%-40%"
})
```

```python
student["Attendance"]=student["Attendance"].str.replace("%","",regex=False)
student["Attendance"]=student["Attendance"].str.split("-").apply(lambda x:(int(x[0])+int(x[1]))/2)
student
```
